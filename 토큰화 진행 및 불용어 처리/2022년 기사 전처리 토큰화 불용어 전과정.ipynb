```python
import pandas as pd
from konlpy.tag import Okt
from tqdm import tqdm, tqdm_pandas

tqdm.pandas()
```

```python
# 강사님으로부터 받은 2022년 기사 데이터 확인
df = pd.read_csv('/content/drive/MyDrive/2023년 10월-11월 프로젝트/카카오 기사데이터/요청데이터/카카오_2022-01-01_2022-12-31.csv')
df
```

```python
# 전처리과정
# 필요없는 속성 제거
df.drop(columns=['Unnamed: 0', 'media','category_sub'],inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

```python
# 날짜 기준으로 정렬
df = df.sort_values(by=['date'])
df.reset_index(drop=True, inplace=True)
df
```

```python
# 전처리 의견 반영함
df = df[df['title'].str.contains(' 카카오 |.카카오 |,카카오 | 카카오,| 카카오.|카카오페이|카카오뱅크|카카오게임즈|카뱅|카겜',na=False)]
df.reset_index(drop=True, inplace=True)
df
```

```python
# 토큰화 및 불용어 처리
def TokenStopword(text):
  okt = Okt()
  okt_tokens = okt.morphs(str(text))
  oktTag = []
  for token in okt_tokens:
    oktTag += Okt().pos(token)
  # 불용어 처리
  stopPos = ['Josa', 'Punctuation', 'Suffix', 'Foreign', 'Alpha', 'Number']
  stopWord = ['연합뉴스','기자','앵커','이데일리','헤럴드','헤럴드경제','뉴스1','세계일보','서울경제'
  ,'파이낸셜','파이낸셜뉴스','이코노미스트','아이뉴스','뉴시스','동아일보','일보','조선일보','중앙일보','디지털뉴스부','인턴기자'
  ,'마이데일리','데일리안','조선비즈','한겨레','지디넷코리아','리포트','한국일보','비즈워치','한', '다', '는', '가', '도', '로']
  stopNoun = ['이', '은', '을', '를'] # '이', '은', '을', '를'은 품사가 'Noun'임
  stopLDA2021 = ['경','즈','널','해야','뒤','식','안','하는','위','초','등','남궁훈','조계','박지윤','조수용','박하선','이경규','유재석'
  ,'','총','분','상','의','오','여','어','주','내년','네','며느','영진','것','익','톡','스케','류영준','강호동','윤영','','뷰','웹','회'
  ,'민수','넘게','할','액','지','터','존','잔','퀵','그','앱','거대','부터','때','인','점','꽃','','남궁','리지','씽씽','배','스','중','손'
  ,'곳','왼쪽','오딘','시','선','달','김일','이틀','우','조계현','전월','답변','황','보사','유희열'] # LDA모델 실행후 제외해야 할 것으로 생각되는 단어를 리스트에 추가해야함!!!
  stopLDA2022 = ['기','이해진','조','전','건','정','소','대','전','홍은택','하는','승','이종호','싶다면','보고','더','년','메','분기','춘'
  ,'료','지','과','고개','근','순','폰','한상혁','각자','주요','예상','속보','두','연내','특징','연','챌','저','없어','부르면','열린','이드'
  ,'붙어','나흘','이어집니다','울','켰다','얘깃','알려주세요','바꿉니다','아냐','듣고','업','년비','씨']
  stopLDA2023 = ['했다','톡','사','아','영','와','구','개월','또','임','형','자','계','합니다','스케','현','장','개','첫','들','오늘','날'
  ,'김','차','겸','명','만','이번','대비','내','하기','부문','커','배재현','어디','부사','탭','무슨','하겠다','듯','이티','식이','땐','곳'
  ,'서','오전','오후','한다','생']
  stopLDAwhole = ['준','아들','딸','텍','기에','농','체','증','받았습니다','이승준','셜','연중','진','쓰일','펍','증','기사','섹션','률','약'
  ,'원','징','플','남','글','그런','홍','허','윤','홍','및','수','지난','했습니다','대해','폭','리','직','엽','이프','그러니까','같아요','그래서'
  ,'뭐','있고요','그때','바로가기']
  word = []
  for tag in oktTag:
    if tag[1] in stopPos: # 무조건 제외
      continue
    elif tag[0] in stopWord:
      continue
    elif tag[0] in stopNoun and (tag[1] == 'Noun' or tag[1] == 'Verb' or tag[1] == 'Adverb'):
      continue
    elif tag[0] in stopLDA2021:
      continue
    elif tag[0] in stopLDA2022:
      continue
    elif tag[0] in stopLDA2023:
      continue
    elif tag[0] in stopLDAwhole:
      continue
    else:
      word.append(tag[0])
  return word
```

```python
df['tokenize_title'] = df['title'].progress_apply(TokenStopword)
df['tokenize_content'] = df['content'].progress_apply(TokenStopword)
```

```python
df.drop(columns=['title','content'],inplace=True)
df.reset_index(drop=True, inplace=True)
df
```
