```python
import pandas as pd
from konlpy.tag import Okt
from tqdm import tqdm, tqdm_pandas

tqdm.pandas()
```

```python
csv_2020 = pd.read_csv('/content/drive/MyDrive/핀인사이트 데이터분석/10 AI 실무 융합/카카오_2020-01-01_2020-12-31.csv')
csv_2020
```

```python
csv_2020.drop(columns=['Unnamed: 0', 'media', 'category', 'category_sub'],inplace=True)
csv_2020.reset_index(drop=True, inplace=True)
csv_2020
```

```python
# 날짜 기준으로 정렬
csv_2020 = csv_2020.sort_values(by=['date'])
csv_2020.reset_index(drop=True, inplace=True)
csv_2020
```

```python
csv_2020 = csv_2020[csv_2020['title'].str.contains(' 카카오 |.카카오 |,카카오 | 카카오,| 카카오.|카카오페이|카카오뱅크|카카오게임즈|카뱅|카겜',na=False)]
csv_2020 = csv_2020[csv_2020['content'].str.contains(' 카카오 |.카카오 |,카카오 | 카카오,| 카카오.|카카오페이|카카오뱅크|카카오게임즈|카뱅|카겜',na=False)]
csv_2020.reset_index(drop=True, inplace=True)
csv_2020
```

```python
# 토큰화 및 불용어 처리
def TokenStopword(text):
  okt = Okt()
  okt_tokens = okt.morphs(str(text))
  oktTag = []
  for token in okt_tokens:
    oktTag += Okt().pos(token)
  # 불용어 처리
  stopPos = ['Josa', 'Punctuation', 'Suffix', 'Foreign', 'Alpha', 'Number']
  stopWord = ['연합뉴스','기자','앵커','이데일리','헤럴드','헤럴드경제','뉴스1','세계일보','서울경제'
  ,'파이낸셜','파이낸셜뉴스','이코노미스트','아이뉴스','뉴시스','동아일보','일보','조선일보','중앙일보','디지털뉴스부','인턴기자'
  ,'마이데일리','데일리안','이데일리','조선비즈','한겨레','지디넷코리아','리포트','한국일보','비즈워치','한', '다', '는', '가', '도', '로', '와', '가']
  stopNoun = ['이', '은', '을', '를', '한', '다', '는', '가', '도', '로', '의', '그', '와']
  stopLDA2019 = ['경','즈','널','해야','뒤','식','안','하는','위','초','등','가', '도', '로', '의', '그', '와', '총','분','상','의','오','여','어','주'
  , '되자', '마자', '들', '에서', '부터', '된다', '그', '와', '등', '후', '였다', '하죠', '찾았다', '#탭에서', '호연지기', '깔렸다', '삼', '김', '이로인해'
  , '포미닛', '이소라', '씨스타', '빠졌음에도', '짤', '잊', '', '놉']
  stopLDA2020 = ['엘리', '쌉싸름해요', '산뜻', '안다', '설리', '도시남녀', '비', '재', '이진욱', '보', '딜로', '나왔다', '송지연', '닉'
  , '면', '간', '및', '큰', '통', '드려야겠네', '롱', '생', '밥', '첫', '요', '온', '즈', '전', '액', '장', '즈', '맛', '민수', '']
  stopLDA2021 = ['경','즈','널','해야','뒤','식','안','하는','위','초','등','남궁훈','조계','박지윤','조수용','박하선','이경규','유재석'
  ,'','총','분','상','의','오','여','어','주','내년','네','며느','영진','것','익','톡','스케','류영준','강호동','윤영','','뷰','웹','회'
  ,'민수','넘게','할','액','지','터','존','잔'] # LDA모델 실행후 제외해야 할 것으로 생각되는 단어를 리스트에 추가해야함!!!
  stopLDA2022 = ['기','이해진','조','전','건','정','소','대','전','홍은택','하는','승','이종호','싶다면','보고','더','년','메','분기','춘'
  ,'료','지','과','고개','근','순','폰','한상혁']
  stopLDA2023 = ['했다','톡','사','아','영','와','구','개월','또','임','형','자','계','합니다','스케']
  word = []
  for tag in oktTag:
    if tag[1] in stopPos: # 무조건 제외
      continue
    elif tag[0] in stopWord:
      continue
    elif tag[0] in stopNoun and (tag[1] == 'Noun' or tag[1] == 'Verb' or tag[1] == 'Adverb'):
      continue
    elif tag[0] in stopLDA2019:
      continue
    elif tag[0] in stopLDA2020:
      continue
    elif tag[0] in stopLDA2021:
      continue
    elif tag[0] in stopLDA2022:
      continue
    elif tag[0] in stopLDA2023:
      continue
    else:
      word.append(tag[0])
  return word
```

```python
csv_2020['tokenize_title'] = csv_2020['title'].progress_apply(TokenStopword)
csv_2020['tokenize_content'] = csv_2020['content'].progress_apply(TokenStopword)
```
