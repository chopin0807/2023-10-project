```python
import pandas as pd
import os
from konlpy.tag import Okt
```

```python
# 강사님으로부터 받은 2022년 기사 데이터 확인
csv_2021 = pd.read_csv('/content/drive/MyDrive/2023년 10월-11월 프로젝트/카카오 기사데이터/요청데이터/카카오_2021-01-01_2021-12-31.csv')
csv_2021
```

```python
# 전처리과정
# 속성 'Unnamed:0'제거
csv_2021.drop(columns=['Unnamed: 0'],inplace=True)
csv_2021.reset_index(drop=True, inplace=True)
csv_2021
```

```python
# 날짜 기준으로 정렬
csv_2021 = csv_2021.sort_values(by=['date'])
csv_2021.reset_index(drop=True, inplace=True)
csv_2021
```

```python
# 기사제목에서 '카카오'라는 키워드가 포함된 행을 추출
# 카카오와 관련 없는 내용이 일부 포함됨 => 필터링 필요
csv_2021[csv_2021['title'].str.contains('카카오', na=False)]
```

```python
# 전처리 의견 반영함
csv_2021 = csv_2021[csv_2021['title'].str.contains(' 카카오 |.카카오 |,카카오 | 카카오,| 카카오.|카카오페이|카카오뱅크|카카오게임즈|카뱅|카겜',na=False)]
csv_2021.reset_index(drop=True, inplace=True)
csv_2021
```

```python
# 토큰화
# 기사제목과 기사내용에서 토큰화 진행함
csv_2021['tokenize(title)'] = csv_2021['title'].apply(lambda x:Okt().morphs(x))
csv_2021
```

```python
csv_2021['tokenize(content)'] = csv_2021['content'].astype('str').apply(lambda x:Okt().morphs(x))
csv_2021
```

```python
csv_2021.to_csv('/content/drive/MyDrive/2023년 10월-11월 프로젝트/카카오 기사데이터/요청데이터/news(2021 토큰화 완료).csv')
```
