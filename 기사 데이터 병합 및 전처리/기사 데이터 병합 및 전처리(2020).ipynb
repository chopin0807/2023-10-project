```python
import pandas as pd
import os
from konlpy.tag import Okt
```

```python
df
```

```python
df_2020 = df[df['title'].str.contains('카카오')]
df_2020
```

```python
df_2020.drop(columns=['Unnamed: 0'],inplace=True)
df_2020
```

```python
# 날짜 기준으로 정렬
df_2020 = df_2020.sort_values(by=['date'])
df_2020.reset_index(drop=True, inplace=True)
df_2020
```

```python
# 기사제목에서 '카카오'라는 키워드가 포함된 행을 추출
# 카카오와 관련 없는 내용이 일부 포함됨 => 필터링 필요
df_2020[df_2020['title'].str.contains('카카오', na=False)]
```

```python
df_2020 = df_2020[df_2020['title'].str.contains(' 카카오 |.카카오 |,카카오 | 카카오,| 카카오.|카카오페이|카카오뱅크|카카오게임즈|카뱅|카겜',na=False)]
df_2020.reset_index(drop=True, inplace=True)
df_2020
```

```python
# 토큰화
# 기사제목과 기사내용에서 토큰화 진행함
df_2020['tokenize(title)'] = df_2020['title'].apply(lambda x:Okt().morphs(x))
df_2020
```

```python
df_2020['tokenize(content)'] = df_2020['content'].astype('str').apply(lambda x:Okt().morphs(x))
df_2020
```

```python
df_2020.drop(columns=['media','category','category_sub','title','content'],inplace=True)
df_2020
```
