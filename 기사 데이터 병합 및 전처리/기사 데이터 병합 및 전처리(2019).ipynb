```python
import pandas as pd
import os
from konlpy.tag import Okt
```

```python
# 토큰화
# 기사제목과 기사내용에서 토큰화 진행함
df_2019['tokenize(title)'] = df_2019['title'].apply(lambda x:Okt().morphs(x))
df_2019
```

```python
df_2019['tokenize(content)'] = df_2019['content'].astype('str').apply(lambda x:Okt().morphs(x))
df_2019
```
