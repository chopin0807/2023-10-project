```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
import csv
```

```python
def KakaoArticle(start_date, end_date): # 각 파라미터에 대한 입력값 형식은 '2023.10.26'으로 str 형식 이여야함
  # csv 파일 쓰기
  f = open("news.csv", "w", newline="", encoding="utf-8")
  wr = csv.writer(f)
  index = ["기사날짜", "기사제목", "네이버 뉴스 url"]
  wr.writerow(index)

  url = 'https://search.naver.com/search.naver?where=news&sm=tab_pge&query=%22%EC%B9%B4%EC%B9%B4%EC%98%A4%22&sort=0&photo=0&field=0&pd=3&ds={}&de={}&cluster_rank=37&mynews=0&office_type=0&office_section_code=0&news_office_checked=&office_category=0&service_area=0&nso=so:r,a:all&start=1'.format(start_date, end_date)
  headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36'}
  res = requests.get(url, headers=headers)
  soup = BeautifulSoup(res.text, 'html.parser')
  csv_value = []
  title = soup.select('div.news_contents > a.news_tit')
  title_list = [] # 기자제목 리스트
  for i in title:
    title_list.append(i['title'])
  naver_article = soup.select('div.info_group')
  article_list = []
  for i in naver_article:
    article_list.append(i)
  # 네이버 기사 링크 리스트로 정리
  naver_article_url = []
  for i in article_list:
    try:
      naver_article_url.append(i.select('a')[1]['href'])
    except:
      naver_article_url.append('url없음')
  # 기사 날짜 추출
  date = soup.select('div.info_group > span')
  date_list = []
  for i in date:
    date_list.append(i.text)
  # 날짜형식(xxxx.xx.xx.)이 나오게 수정
  article_date = []
  for i in date_list:
    p = re.compile('[0-9][0-9][0-9][0-9].[0-9][0-9].[0-9][0-9].')
    t = re.compile('[0-9가-힣]* 전')
    if p.match(i) != None or t.match(i) != None:
      article_date.append(i)

  for i in range(len(title_list)):
    temp = []
    temp.append(date_list[i])
    temp.append(title_list[i])
    temp.append(naver_article_url[i])
    csv_value.append(temp)
  print(csv_value)
  for i in csv_value:
    wr.writerow(i)
  f.close()
```

```python
KakaoArticle('2023.01.01','2023.10.26') # 코렙 파일시스템에서 /content/news.csv로 생성됨
```
