```python
# 2021.01 데이터 합치기
file_names = glob('카카오 기사데이터/최준호(2021)/2021.01/*.csv') # 폴더 내의 모든 csv 파일 호출
total = pd.DataFrame()

for file_name in file_names:
    file_name = file_name.replace('\\', '/')
    temp = pd.read_csv(file_name)
    total = pd.concat([total, temp]) # 전체 데이터프레임에 추가하여 넣는다.
    
total.to_csv('카카오 기사데이터/최준호(2021)/news(2021.01).csv')
```

```python
# 파일 확인
df_2021_01 = pd.read_csv('카카오 기사데이터/최준호(2021)/news(2021.01).csv')
df_2021_01
```

```python
# 전처리과정
# 속성 'Unnamed:0'제거
df_2021_01.drop(columns=['Unnamed: 0'],inplace=True)
df_2021_01
```

```python
# 속성 '기사내용'에서 '<네이버 기사 없음>'이나 '<구조 다름>'모두 삭제
find = df_2021_01[df_2021_01['기사내용'] == '<네이버 기사 없음>'].index
df_2021_01.drop(find, inplace=True)
df_2021_01
```

```python
find = df_2021_01[df_2021_01['기사내용'] == '<구조 다름>'].index
df_2021_01.drop(find, inplace=True)
df_2021_01
```
